```python
#improting librararies
import pandas as pd
import warnings
warnings.filterwarnings('ignore', 'The spaces in these column names will not be changed.')
import sqlite3
```

```python
#creating data frame from CSV file
data = pd.read_csv("Real_Estate_Sales_2001-2020_GL.csv", low_memory=False)
```

```python
data
```

```python
#connecting to SQLite3/Creating Table from Dataframe
connection = sqlite3.connect("real_estate.db")

data.to_sql("real_estate_table", connection, if_exists="replace")
```

```python
#Attaching Query to dataframe ing sql Table format
df = pd.read_sql_query("SELECT * FROM real_estate_table", connection)
```

```python
df
```

```python
#This query generates all records from 2015 and above. Using the List Year column
year_q = pd.read_sql_query("SELECT * FROM real_estate_table WHERE [List Year] >= 2015 ORDER BY [List Year] ASC", connection)
```

```python
year_q
```

```python
#converting the query into a table
df[df['List Year'] >= 2015].to_sql('year_q', connection, if_exists="replace")
```

```python
year_q
```

```python
#This query generates all "condo" records from Property Type Column and Residential Type Column
condo_q = pd.read_sql_query("SELECT * FROM real_estate_table WHERE [Property Type] == 'Condo' AND [Residential Type] == 'Condo' ORDER BY [List Year] ASC", connection)
```

```python
condo_q
```

```python
#Again turning that into the query from above into a table
df[(df['Property Type'] == 'Condo') & (df['Residential Type'] == 'Condo')].to_sql('condo_q', connection, if_exists="replace")
```

```python
condo_q
```

```python
#Using INNER JOIN to merge year_q table and condo_q table on List Year Column to give me all "condo" data from 2105 and above
#side note had to limit to 900,000 because dataframe was too big
combined_query = "SELECT * FROM (SELECT * FROM real_estate_table WHERE [List Year] >= 2015) AS year_q INNER JOIN (SELECT * FROM real_estate_table WHERE [Property Type] = 'Condo' AND [Residential Type] = 'Condo') AS condo_q ON year_q.[List Year] = condo_q.[List Year] LIMIT 900000"
combined_df = pd.read_sql_query(combined_query, connection, if_exists="replace")
```
